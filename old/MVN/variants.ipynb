```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Univariate Normal Distribution
mu, sigma = 0, 1  # Mean and standard deviation
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, mu, sigma)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pdf, 'r-', lw=2)
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
plt.show()
```

```python
plt.figure()
plt.plot(x, pdf, label=f'N({mu}, {sigma}^2)')
plt.fill_between(x, pdf, where=(x <= 1), color='gray', alpha=0.5)
plt.title('Univariate Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
p_x_leq_1 = stats.norm.cdf(1, mu, sigma)
print(f'P(X <= 1) for N(0,1): {p_x_leq_1:.4f}')
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'N({mu}, {sigma}^2)')
```

```python
p_x_geq_1 = 1 - stats.norm.cdf(1, mu, sigma)
print(f'P(X >= 1) for N(0,1): {p_x_geq_1:.4f}')
```

```python
# Bivariate Normal Distribution
mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]  # Covariance matrix
x, y = np.mgrid[-3:3:.1, -3:3:.1]
pos = np.dstack((x, y))
rv = stats.multivariate_normal(mean, cov)
z = rv.pdf(pos)
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('Bivariate Normal Distribution')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Density')
plt.show()
```

```python
p_xy_leq_1 = rv.cdf([1, 1])
print(f'P(X <= 1, Y <= 1) for Bivariate Normal: {p_xy_leq_1:.4f}')
```

```python
p_xy_geq_1 = 1 - rv.cdf([1, 1])
print(f'P(X >= 1, Y >= 1) for Bivariate Normal: {p_xy_geq_1:.4f}')
```

```python
# Multivariate Normal Distribution (3D)
mean = [0, 0, 0]
cov = [[1, 0.5, 0.3], [0.5, 1, 0.4], [0.3, 0.4, 1]]
rv_multi = stats.multivariate_normal(mean, cov)
p_xyz_leq_1 = rv_multi.cdf([1, 1, 1])
print(f'P(X <= 1, Y <= 1, Z <= 1) for Multivariate Normal: {p_xyz_leq_1:.4f}')
```

```python
# Generate samples from the 3D multivariate normal
samples = rv_multi.rvs(size=500)

# Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2], alpha=0.5, c='b', marker='o')

# Labels and title
ax.set_title('3D Scatter Plot of Multivariate Normal Distribution')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
