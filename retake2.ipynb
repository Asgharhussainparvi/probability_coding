```python
correlation_matrix = df[['selling_price', 'km_driven', 'year']].corr()
print(correlation_matrix)

sns.pairplot(df[['selling_price', 'km_driven', 'year']])
plt.show()
```

```python
mu, std = stats.norm.fit(df['selling_price'])

probability_above_1000000 = 1 - stats.norm.cdf(1000000, mu, std)
print(f"Probability of a car being priced above 10,00,000: {probability_above_1000000}")

# Explanation
print("This probability represents the likelihood that a randomly selected car from the dataset will have a selling price greater than 10,00,000.")
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
print(df.describe())

df[['selling_price', 'km_driven', 'year']].hist(bins=30, figsize=(10, 7))
plt.show()

df[['selling_price', 'km_driven', 'year']].plot(kind='box', subplots=True, layout=(1, 3), figsize=(15, 5))
plt.show()
```

```python
mean_vector = df[['selling_price', 'km_driven', 'year']].mean().values
cov_matrix = df[['selling_price', 'km_driven', 'year']].cov().values

print(f"Mean Vector: {mean_vector}")
print(f"Covariance Matrix: {cov_matrix}")


# Heatmap of the covariance matrix
sns.heatmap(df[['selling_price', 'km_driven', 'year']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
hypothetical_car = [800000, 50000, 2016]

# Calculate the probability density
prob_density = stats.multivariate_normal.pdf(hypothetical_car, mean=mean_vector, cov=cov_matrix.T)
print(f"Probability Density for the hypothetical car: {prob_density}")
```
